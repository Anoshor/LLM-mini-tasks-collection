```python
from bing_image_downloader import downloader
import os
import time

# Define the classes and the number of images per class
classes = ["bag", "pen", "laptop", "phone", "bed", "vehicle", "chair", "bottle", "shoes", "table"]
num_images_per_class = 200
output_dir = "dataset"


os.makedirs(output_dir, exist_ok=True)

for class_name in classes:
    print(f"Downloading images for class: {class_name}")
    downloader.download(
        class_name,
        limit=num_images_per_class,
        output_dir=output_dir,
        adult_filter_off=True,
        force_replace=False,
        timeout=60
    )
    print(f"Successfully downloaded images for class: {class_name}")

print("Image downloading completed.")
```

```python
from datasets import Dataset, DatasetDict
import os
import shutil
from PIL import Image
from config import HUGGINGFACE_TOKEN
```

```python
classes = ["bag", "pen", "laptop", "phone", "bed", "vehicle", "chair", "bottle", "shoes", "table"]
output_dir = "dataset"
```

```python
file_paths = []
labels = []
```

```python
# Iterate through the dataset directory and collect file paths and labels
for class_name in classes:
    class_dir = os.path.join(output_dir, class_name)
    for file_name in os.listdir(class_dir):
        file_path = os.path.join(class_dir, file_name)
        try:
            # Verify that the file is a valid image
            Image.open(file_path).verify()
            file_paths.append(file_path)
            labels.append(class_name)
        except (IOError, SyntaxError) as e:
            print(f"Skipping corrupted image: {file_path}")
```

```python
# Create a Dataset
data = {"image": file_paths, "label": labels}
dataset = Dataset.from_dict(data)
```

```python
# Split the dataset into training and validation sets
train_test_split = dataset.train_test_split(test_size=0.2)
```

```python
dataset_dict = DatasetDict({"train": train_test_split['train'], "test": train_test_split['test']})
```

```python
dataset_dict.push_to_hub("Anoshor/fatima-fellowship")

print("Dataset uploaded to Hugging Face Hub.")
```
